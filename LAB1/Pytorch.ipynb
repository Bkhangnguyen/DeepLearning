```python
import torch

x = torch.tensor(1.0, requires_grad=True)
y = 5*x**5 + 6*x**3 + 3*x + 1
y.backward()

print("Giá trị x : ", x.item())
print("Độ dốc tại điểm đó là", x.grad.item())
```

##### BTVN 2: Tạo một tensor ban đầu có giá trị là 2. Định nghĩa hàm số và tính gradient
            y = x ^ 3 + 2x ^ 2 + 5x + 1             

- Hãy tính dy/dx tại giá trị của x

- Dùng phương pháp Gradient Descent với Learning Rate Alpha = 0.1 để cập nhật giá trị x trong 10 vòng lặp 


```python
import torch

x = torch.tensor(2.0, requires_grad=True)
y = x**3 + 2*x**2 + 5*x + 1
y.backward()

print("Giá trị x ban đầu : ", x.item())
print("Gradient dy/dx tại x : ", x.grad.item())
```

```python
learning_rate = 0.1
input_tensor = torch.tensor(2.0, requires_grad=True)

print("Biểu đồ giảm Gradient: ")
print("=" * 40)

for iteration in range(10):
    if input_tensor.grad is not None:
        input_tensor.grad.zero_()
    
    output = input_tensor**3 + 2*input_tensor**2 + 5*input_tensor + 1
    output.backward()
    
    with torch.no_grad():
        input_tensor -= learning_rate * input_tensor.grad
    
    print(f"Lần {iteration+1:2d}: giá trị x = {input_tensor.item():.6f}, đạo hàm = {input_tensor.grad.item():.6f}")
```

##### BTVN 3: Tạo một tập dữ liệu giả lập với x là số giờ học (ngẫu nhiên từ 1 - 10) và y là số điểm tính theo công thức: 
        
            y = 3x + 5 + noise             

**Ghi chú:** --> noise là một giá trị ngẫu nhiên nhỏ

1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad = True
2. Tính MSE
3. Tính Gradient
4. Cập nhật tham số w và b bằng Gradient Descent với learning rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình.

```python
import torch

# Tạo x: số giờ học từ 1 đến 10
x = torch.rand(20, 1) * 9 + 1
print("Số giờ học (x) - 5 mẫu đầu:")
print(x[:5])

# Tạo noise nhỏ
noise = torch.randn(20, 1) * 1.0

# Tạo y theo công thức: y = 3x + 5 + noise
y = 3 * x + 5 + noise
print("\nĐiểm số (y) - 5 mẫu đầu:")
print(y[:5])
```

```python
# Tạo w và b ngẫu nhiên
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print(f"w = {w.item():.4f}")
print(f"b = {b.item():.4f}")
print(f"\nMô hình ban đầu: y = {w.item():.4f} * x + {b.item():.4f}")
print("Mô hình thực tế: y = 3.0 * x + 5.0")
```

```python
# Tính dự đoán
y_pred = w * x + b

# Tính MSE loss
loss = torch.mean((y_pred - y) ** 2)

print("Ví dụ 3 mẫu đầu:")
for i in range(3):
    print(f"x={x[i].item():.1f} -> y_thực={y[i].item():.2f}, y_dự_đoán={y_pred[i].item():.2f}")

print(f"\nMSE Loss = {loss.item():.4f}")
```

```python
# Tính gradient
loss.backward()

print(f"Gradient của w: {w.grad.item():.4f}")
print(f"Gradient của b: {b.grad.item():.4f}")

# Giải thích đơn giản
if w.grad.item() > 0:
    print("-> w đang quá lớn, cần giảm w")
else:
    print("-> w đang quá nhỏ, cần tăng w")
```

```python
# Learning rate
alpha = 0.01

# Lưu giá trị cũ
w_cu = w.item()
b_cu = b.item()

# Cập nhật
with torch.no_grad():
    w -= alpha * w.grad
    b -= alpha * b.grad

# Xóa gradient
w.grad.zero_()
b.grad.zero_()

print(f"Trước cập nhật: w={w_cu:.4f}, b={b_cu:.4f}")
print(f"Sau cập nhật:  w={w.item():.4f}, b={b.item():.4f}")
print(f"w thay đổi: {w.item() - w_cu:.4f}")
print(f"b thay đổi: {b.item() - b_cu:.4f}")
```

```python
# Khởi tạo lại tham số
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
alpha = 0.01

print(f"Bắt đầu: w={w.item():.4f}, b={b.item():.4f}")
print("\nTiến trình:")
print("-" * 50)

for i in range(100):
    # Tính dự đoán
    y_pred = w * x + b
    
    # Tính loss
    loss = torch.mean((y_pred - y) ** 2)
    
    # Tính gradient
    loss.backward()
    
    # Cập nhật
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
    
    # Xóa gradient
    w.grad.zero_()
    b.grad.zero_()
    
    # In kết quả
    if (i + 1) % 20 == 0:
        print(f"Lần {i+1:3d}: Loss={loss.item():.4f}, w={w.item():.4f}, b={b.item():.4f}")

print("-" * 50)
print(f"\nKết quả cuối:")
print(f"w = {w.item():.4f} (gần 3.0)")
print(f"b = {b.item():.4f} (gần 5.0)")
print(f"Loss = {loss.item():.4f}")

# Kiểm tra
print(f"\nKiểm tra:")
x_test = 4.0
y_thuc = 3 * x_test + 5
y_du_doan = w.item() * x_test + b.item()
print(f"Với x={x_test}:")
print(f"  Thực tế: y = {y_thuc:.2f}")
print(f"  Dự đoán: y = {y_du_doan:.2f}")
print(f"  Sai số: {abs(y_thuc - y_du_doan):.2f}")
```

# Sự khác biệt này được giải thích như sau:

**1.** Nguyên nhân xuất phát từ phương pháp tạo tensor khác nhau trong PyTorch.

**2.** Khi sử dụng `torch.from_numpy()`, PyTorch tạo tensor bằng cách tham chiếu đến cùng một vùng bộ nhớ với mảng NumPy. Do đó, mọi thay đổi trên mảng NumPy sẽ ngay lập tức phản ánh trên tensor.

**3.** Ngược lại, với `torch.tensor()`, PyTorch tạo ra một tensor mới có bản sao độc lập của dữ liệu. Tensor này không chia sẻ bộ nhớ với mảng NumPy gốc, nên việc thay đổi mảng NumPy không ảnh hưởng đến tensor đã tạo.
